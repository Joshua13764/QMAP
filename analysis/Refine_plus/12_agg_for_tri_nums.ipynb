```python
combined_db_export_pool = Path("combined_db_export_pool")
combined_db_pass_00_export_folder: Path = combined_db_export_pool / "pass_00"
export_pool = Path("agg_export_pool")
```

```python
# pl.scan_parquet(export_folder).group_by("boulder_id").agg().collect()

filter_df = pl.scan_parquet(combined_db_pass_00_export_folder).filter(pl.col("boulder_id") < 40).collect()
filter_df.head()
```

```python
cols_to_group: list[str] = [col_name for col_name in filter_df.columns if "tri_num" in col_name]

for col in cols_to_group:
    print(
        filter_df.select(col).group_by(col).agg(pl.len().alias("facet_alpha"))
    )
```

```python
for col in cols_to_group:
    print(f"Joining for col {col}")
    df = pl.scan_parquet(combined_db_pass_00_export_folder).select(col).group_by(
        col
    ).agg(pl.len().alias("facet_alpha")).collect(engine="streaming")
    df.write_parquet(export_pool / f"{col}.parquet")
```
